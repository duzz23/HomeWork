```python
df_top = df.filter(items = ['Surgery?','Age','Rectal_temperature','Abdomen','Abdominocentesis_appearance','type_of_lesion1','type_of_lesion2','type_of_lesion3'])
df_top.head()

```

```python
for col in df_top.columns:
    
    pct_missing = df_top[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

# Проверяем на выбросы 

```python
df_new = df_top.fillna(0)
df_new
```

```python
df_new.describe()
```

# 1. Замечны выбросы в строке: 
Age(значения 9 не должно быть), Abdomen (значения 5 не должно быть)
# 2. Работаем с выбросами 

```python
q1 = df_new.Age.quantile(0.25)
q3 = df_new.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_new[df_new.Age.between(lower_bound, upper_bound, inclusive=True)].sort_values('Surgery?')
remove_outliers
```

```python
df_blowout = pd.concat([df_new, remove_outliers]).drop_duplicates(keep=False)
df_blowout
```

# Выброс
Вот мы и нашли выброс по возрасту, если мы его удалим потеряем много данных, как вариант можно заменить на предыдушее значение. 

```python
df = pd.DataFrame(df_new)
df = df.replace(to_replace=9, value=2, regex=True)
df = df.replace(to_replace=5.0, value=0.0, regex=True)
```
