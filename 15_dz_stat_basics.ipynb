# Ищу выброс в колонке Rectal_temperature, судя по всему есть минемальное температура и максимальная 0 и 40 которое требет изменений 


```python
q1 = df_top['Rectal_temperature'].quantile(0.25)
q3 = df_top['Rectal_temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_top[df_top['Rectal_temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Surgery?')
remove_outliers
```

```python
df_drop = pd.concat([df_top, remove_outliers]).drop_duplicates(keep=False)
df_drop
```

```python
#Так как в этих  данных толком нет информации была ли операция у животного и нет температуры: думаю что эти данные мне не нужны для дальнейшей работы 
#удалю те значения в которых как минимум 6/8 значений заполнено
df11 = df_drop.dropna(thresh=6)
df11
```

# Проверяем на выбросы 

```python
df_new = df11.fillna(0)
df_new
```

```python
q1 = df_new.Age.quantile(0.25)
q3 = df_new.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_new[df_new.Age.between(lower_bound, upper_bound, inclusive=True)].sort_values('Surgery?')
remove_outliers
```

```python
df_blowout = pd.concat([df_new, remove_outliers]).drop_duplicates(keep=False)
df_blowout
```
