```python
import pandas as pd
import numpy as np
```

```python
log_hw = pd.read_csv('/Users/macbookpro/Desktop/Netology/Home_Work/Python_12_pivot_and_str/visit_log.csv', sep=';')
log_hw.head()
```

```python
source_1 = ['paid', 'email', 'yandex', 'google']
for traffic_source in source_1:
    log_hw['source_type'] = log_hw['traffic_source']
    log_hw.loc[(log_hw.region == 'Russia') & (log_hw.traffic_source == 'paid') | (log_hw.traffic_source == 'email'), 'source_type'] = 'Ad'
    log_hw.loc[(log_hw.region != 'Russia') & (log_hw.traffic_source == 'paid') | (log_hw.traffic_source == 'email'), 'source_type'] = 'other'   
    log_hw.loc[log_hw.traffic_source.isin (['google', 'yandex']) , 'source_type'] = 'organic'
log_hw.head()
```

# Задание 3

```python
import pandas as pd
```

```python
movies_hw = pd.read_csv('/Users/macbookpro/Desktop/Netology/Home_Work/ratings.csv')
movies_hw["ratingCount"] = movies_hw.groupby('userId').rating.count()
movies_hw.head(5)
```

```python
movies_100 = movies_hw.loc[movies_hw.ratingCount > 100]
movies_100['real_time'] = pd.to_datetime(movies_100['timestamp'], unit='s')
movies_max = movies_100.groupby('userId').max()[['real_time']]
movies_min = movies_100.groupby('userId').min()[['real_time']]
movies_result = movies_max - movies_min
movies_result.head()
```
